### Now I am going to treat each month as its own cohert, from there I can work out the Churn rate/percentage per month.

```python
yes_churn_perc = []

for month in range(1, 73):
    
    perc = 100 * len(df[(df['tenure'] == month) & (df['Churn'] == 'Yes')]) / len(df[(df['tenure'] == month)])
    yes_churn_perc.append(perc)
```

```python
percentage_df = pd.DataFrame(yes_churn_perc, index=range(1, 73), columns=['% of Customer Churn'])
```

```python
percentage_df = percentage_df.reset_index().rename(columns={'index': 'Month On Contract'})
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=percentage_df, x='Month On Contract', y='% of Customer Churn');
```

### I'd like to make broader cohort groups, for example '0 - 12 Months' etc for tunure.

```python
bins = [0, 12,  24, 48, df['tenure'].max()]

labels = ['0-12 Months', '12-24 Months', '24-48 Months', 'Over 48 Months']

df['Tenure Cohort'] = pd.cut(df['tenure'], bins=bins, labels=labels, include_lowest=True)
```

## Predictive modelling. 

### I'll start with a single decision tree and also I will need to sort the data so that is it ready to be used for machine learning.

```python
X = df.drop(['customerID', 'Tenure Cohort', 'Churn'], axis=1)
```

```python
y = df['Churn']
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
dt = DecisionTreeClassifier()
```

```python
param_grid = {
    'max_depth': [None, 6, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

grid_search_dt = GridSearchCV(dt, param_grid=param_grid)
```

```python
grid_search_dt.fit(X_train, y_train)
```

### Now I'll predict off this model and check the metrics.

```python
dt_preds = grid_search_dt.predict(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
```

```python
print(classification_report(y_test, dt_preds))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, dt_preds);
```

### The best parameters and feature importance.

```python
grid_search_dt.best_params_
```

```python
plt.figure(figsize=(14,8))
plt.title('Importance of Each Column')

feature_importances = grid_search_dt.best_estimator_.feature_importances_

feature_df = pd.DataFrame(feature_importances, index=X_train.columns, columns=['importance']).sort_values('importance')

sns.barplot(data=feature_df, x='importance', y=feature_df.index);
```

## AdaBoosted Tree

### Same again but now i will use AdaBoost and gridsearch.

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.5, 1]
}
```
